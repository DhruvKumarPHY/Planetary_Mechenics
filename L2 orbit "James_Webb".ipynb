```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
import math as m
mod = m.fabs
```

```python
# Define our constants
M = 1.989* 10**30 # mass of sun in Kg
m = 5.972 * 10**24 # mass Earch in Kg
r = 151.65 *10**6 # distance of Earth from Sun in Km
```

```python
def f(d):
    return M*d**2 + (r + d)**2 * (m - (M/(r**3)) * d**2 *(r + d))
```

```python
def f_(f,d):
    h = 0.0000001
    return (f(d+h)-f(d))/h
```

```python
def f__(f,d):
    h = 0.0000001
    return (f_(f,d+h)-f_(f,d))/h
```

```python
def nr(f,d0,e =0.0000001):
    d1 = None
    N = 1000
    n =1
    print("d0 = ",d0)
    
    while(n<N and mod(f(d0))>e):
        h = -f(d0)/f_(f,d0)
        d1 = d0 + h
        print(n,d1)
        d0 = d1
        n = n+1
    return d0 if f(d0)<e else "Max Interation Exceeded"
```
